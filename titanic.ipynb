```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# menampilkan dataset titanic ke dalam dataframe
data = pd.read_csv('../csv/titanic.csv')
df = pd.DataFrame(data)
df
```

```python
# digunakan untuk mengetahui baris mana saja yang terdapat missing value
df.isna()
```

```python
# digunakan untuk menjumlahan data yang terdapat NaN
df.isna().sum()
```

```python
# mengecek data age
min = df['age'].min()
max = df['age'].max()
mean = df['age'].mean()
median = df['age'].median()
print(min)
print(max)
print(mean)
print(median)
```

```python
# mengisi kolom age dengan mean
df['age'] = df['age'].fillna(df['age'].mean())
df.isna().sum()
```

```python
df['fare'] = df['fare'].fillna(df['fare'].mean())
df
```

```python
df['embarked'] = df['embarked'].fillna('unknown')
df.isna().sum()
```

```python
# mendapatkan data object berdasarkan nilai tertentu
df[df.embarked == 'unknown'].embarked
```

```python
df['cabin'] = df['cabin'].fillna('unknown')
df
```

```python
df
```

```python
# mendapatkan data survived berdasarkan kolom sex
df.groupby('sex')['survived'].agg(['count', 'mean', 'median'])
```

```python
# menampilkan data dengan query
df.query('age < 10')
```

```python
# menyeleksi data
df[(df['survived'] == 1) & (df['age'] == 20)]
```

```python
# agregasi beberapa kolom
# dibuat berdasarkan pclass
# menghitung setiap orang yang berada pada class 1,2,3
# menghitung mean, min, max berdasarkan usia pada suatu class
# menghitung rata-rata survived suatu class
agg_col = df.groupby('pclass').agg({'pclass':'count',
                                    'age':['mean', 'min', 'max'],
                                    'survived':'mean'})
agg_col
```

```python
# merubah nama kolom
agg_col = df.groupby('pclass').agg(n_pass=('pclass', 'count'),
                                   avg_age=('age', 'mean'),
                                   min_age=('age', 'min'),
                                   max_age=('age', 'max'),
                                   survived=('survived','mean'))
agg_col
```
